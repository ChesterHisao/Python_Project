```python
#The purpose of this fuction is about quicker revising queries.
#Full version is for numerical with range fuction, short/other is for char.
def get_into_string(feature_name, version='full', option=''):
  
  if version == 'full':
    result= f"""Which Student {feature_name} do you want? 
  {option}
  If you want to search mutltiple {feature_name} please use ',' to separate
  If you want to search all {feature_name} begin with X, please type X% 
  If you want to search all {feature_name} ending with X, please type %X 
  If you want to search all {feature_name} after X, please type >X 
  If you want to search all {feature_name} after or equal to X, please type >=X   
  If you want to search all {feature_name} before X, please type <X 
  If you want to search all {feature_name} before or equal to X, please type <=X   
  If you want to search all {feature_name} between X and Y, please type X~Y \n"""

  else:
    result= f"""Which Student {feature_name} do you want? 
  {option}
  If you want to search mutltiple {feature_name} please use ',' to separate
  If you want to search all {feature_name} begin with X, please type X% 
  If you want to search all {feature_name} ending with X, please type %X \n"""


  return result
```

```python
#Distinguish the type of query, start with should input str
restriction_query_clean = []
if user_query != 7:  
  user_column = list(df.columns)[user_query-1]


  if '%' in restriction_query:
    if '%' in restriction_query.strip()[0]:
      query_method = 'end_with'
      restriction_query_clean = restriction_query.strip()[1:].strip().lower()

    else:
      query_method = 'begin_with'
      restriction_query_clean = restriction_query.strip()[0:-1].strip().lower()

  elif '>=' in restriction_query:
     query_method = 'larger_equal'
     restriction_query_clean = int(restriction_query.strip()[2:].strip())

  elif '<=' in restriction_query:
    query_method = 'smaller_equal'
    restriction_query_clean = int(restriction_query.strip()[2:].strip())

  elif '>' in restriction_query:
     query_method = 'larger'
     restriction_query_clean = int(restriction_query.strip()[1:].strip())

  elif '<' in restriction_query:
    query_method = 'smaller'
    restriction_query_clean = int(restriction_query.strip()[1:].strip())


  elif '~' in restriction_query:
    query_method = 'between'
    restriction_query_split_temp = restriction_query.split('~')
    restriction_query_clean.append(int(restriction_query_split_temp[0].strip()))
    restriction_query_clean.append(int(restriction_query_split_temp[1].strip()))

  #Transfer the result from string to list
  else:
    query_method = 'list'
    restriction_query_split = restriction_query.split(',')

    #Num and char variables should have different way to transfer 
    for i in restriction_query_split:
      
      # Correct user_column type but wrong 
      try:
        if user_column in nvar_list:
          i = int(i.strip())
        else:
          i = i.strip().lower()

        restriction_query_clean.append(i)

      except:
        print('This is not a number!')

  
```

```python

def query_fuction(user_query, restriction_query_clean, query_method, user_query_type, user_query_group=0, display_user_query_list=None):

  #Display all rows
  pd.set_option('display.max_rows', None)

  if user_query == 7:
    result_df = df

    #If user chose none, we don't need the belowed fuctions
    #return 


  else:
    user_column = list(df.columns)[user_query-1]
        #Slicing: finding truth and then only display it
    if query_method == 'list':
      result_df = df[df[user_column].isin(restriction_query_clean)]    

    elif query_method == 'begin_with':
      result_df = df[df[user_column].str.startswith(restriction_query_clean)]

    elif query_method == 'end_with':
      result_df = df[df[user_column].str.endswith(restriction_query_clean)]

    elif query_method == 'larger_equal':
      result_df = df[df[user_column]>=(restriction_query_clean)]

    elif query_method == 'smaller_equal':
      result_df = df[df[user_column]<=(restriction_query_clean)]

    elif query_method == 'larger':
      result_df = df[df[user_column]>(restriction_query_clean)]

    elif query_method == 'smaller':
      result_df = df[df[user_column]<(restriction_query_clean)]

    elif query_method == 'between':
      result_df = df[df[user_column].between(restriction_query_clean[0],restriction_query_clean[1])]


  if user_query_type == 1:
    print(result_df[display_user_query_list])

  else:       
    group_by_column = list(df.columns)[user_query_group-1]
    sr2 = result_df.groupby(group_by_column).ID.count()
    sr2 = sr2.rename('Count')
    sr3 = (sr2*100)/sr2.sum()
    sr3 = sr3.rename('%')

    df2 = pd.concat([sr2, sr3], axis=1)
    
    # average horsepower per engine type
    print(df2)

```

```python
query_fuction(user_query, restriction_query_clean, query_method, user_query_type, user_query_group, display_user_query_list)
```
